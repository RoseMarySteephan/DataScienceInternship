```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv('world bank.csv', skiprows=4)
df.head()
```

```python
df_clean = df[['Country Name', '2022']].dropna()
df_clean.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_clean['2022'], bins=30, kde=True, color='skyblue')
plt.title('Population Distribution of Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```

```python
top10 = df_clean.sort_values(by='2022', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', data=top10, palette='viridis')
plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
